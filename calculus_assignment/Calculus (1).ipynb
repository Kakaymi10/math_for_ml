# Task 1: Write a dynamic function to find the derivative of any function f(x)

````python
import sympy as sp

from sympy import symbols, diff, simplify
def dy_dx(expression, variable='x'):
  x = symbols(variable)
  derivative = diff(expression, x)
  return simplify(derivative)

expression = '4*x**2 + 3*x + 2'
derivative_result = dy_dx(expression)
print(f"The derivative of {expression} is: {derivative_result}")
````

#Task 3: Plot a graph of the quadratic Equation

````python
import numpy as np
import matplotlib.pyplot as plt

# Plot the function and its derivative
x_values = np.linspace(0, 4, 100)  # Generate 100 x values from 0 to 4
y_values = [f(x) for x in x_values]
derivative_values = [dy_dx(f(x)) for x in x_values]

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(x_values, y_values, label='f(x)')
plt.title('Quadratic Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x_values, derivative_values, label="f'(x)")
plt.title('Derivative of Quadratic Function')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

plt.tight_layout()
plt.show()
````

# Task 4: Create separate array of Minimas and maximas

````python
from scipy.signal import argrelextrema
# Find local minimas and maximas using argrelextrema
minima_indices = argrelextrema(np.array(y_values), np.less)[0]
maxima_indices = argrelextrema(np.array(y_values), np.greater)[0]

# Extract x values corresponding to minimas and maximas
minima_x_values = x_values[minima_indices]
maxima_x_values = x_values[maxima_indices]

print(f'The array of minimas: {minima_x_values} and the array of maximas: {maxima_x_values}')
````

Quadratic Function with No Global Extrema:

- If the quadratic function opens upwards (leading coefficient
a is positive), it will have a global minimum, but it may not have a global maximum.

- If the quadratic function opens downwards (leading coefficient

a is negative), it will have a global maximum, but it may not have a global minimum.


````python
from scipy.optimize import minimize_scalar
# Define the function to be minimized and maximized
objective_function = lambda x: -f(x)  # Negative of the function for maximization

# Find the global maximum
result_max = minimize_scalar(objective_function, bounds=(0, 4), method='bounded')
global_maxima = (result_max.x, -result_max.fun)  # (x, y)

# Find the global minimum
result_min = minimize_scalar(f, bounds=(0, 4), method='bounded')
global_minima = (result_min.x, result_min.fun)  # (x, y)

print(f'The global maxima is {global_maxima}')
print(f'The global minima is {global_minima}')
````
